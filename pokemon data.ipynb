```python
import pandas as pd
#importar pandas como pd
```

```python
df = pd.read_csv("C:\\Users\\User\\Downloads\\pokemon_data.csv", index_col = 0)
#ler o arquivo, colocar a coluna # como index
```

```python
df.head()
#verificar o formato dos dados
```

```python
df["Stats AVG"] = df[["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]].mean(axis=1)
#media de status na coluna "Stats AVG"
```

```python
df.loc[df["Legendary"] == True].sort_values(by="Stats AVG", ascending=False).head(10)
#10 pokemon lendários com maiores médias de status
```

```python
df.loc[df["Legendary"] == False].sort_values(by="Stats AVG", ascending=False).head(10)
#10 maiores medias sem ser lendários
```

```python
'''vamos supor que será necessário criar um time com 6 tipos 1 diferentes, com no máximo 1 lendário,
qual a maior média dos status possível para o time?
'''
max_type1 = df[df["Legendary"] == True].groupby("Type 1")[["Stats AVG"]].max()
#cria a variavel com os maiores "Stats AVG" lendários para cada "Type 1"
max_type1.sort_values(by ="Stats AVG", ascending = False)
#mostra por ordem de maior para menor 
```

```python
max_type1_not_legendary = df[df["Legendary"] == False].groupby("Type 1")[["Stats AVG"]].max()
max_type1_not_legendary.sort_values(by ="Stats AVG", ascending = False)
#repetindo o processo, porém com apenas os maiores não lendários, percebemos que os 5 tipos 1 com o pokemon de maior média
#de status são: Dragão, Ferro, Pedra, Normal, Água
```
